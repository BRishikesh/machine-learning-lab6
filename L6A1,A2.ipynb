```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load the data
df = pd.read_excel('embeddingsdata.xlsx')
Embed_264 = df['embed_264']
Embed_265 = df['embed_265']

# Create the scatter plot
plt.scatter(Embed_264,Embed_265)
# Set the axis labels
plt.xlabel("embed_264")
plt.ylabel("embed_265")

# Add a title
plt.title("Scatter plot of embed_264 vs. embed_265")

# Show the plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

# Load the dataset
df = pd.read_excel("embeddingsdata.xlsx")

# Assume 'embed_264' is the independent variable
independent_variable = df['embed_264']
dependent_variable = df['embed_265']

# Create a linear regression model
model = LinearRegression()
model.fit(independent_variable.values.reshape(-1, 1), dependent_variable.values)

# Predict the values
predicted_values = model.predict(independent_variable.values.reshape(-1, 1))

# Calculate mean squared error
mse = mean_squared_error(dependent_variable, predicted_values)
print(f'Mean Squared Error: {mse}')

# Create a scatter plot with regression line
plt.figure(figsize=(8, 6))
plt.scatter(independent_variable, dependent_variable, color='blue', alpha=0.5)
plt.plot(independent_variable, predicted_values, color='red', label='Regression Line', linewidth=2)

# Add labels and title
plt.xlabel('embed_264')
plt.ylabel('embed_265')
plt.title('Linear Regression Model: embed_264 vs embed_265')

# Add legend
plt.legend()

# Show plot
plt.grid(True)
plt.show()

```
